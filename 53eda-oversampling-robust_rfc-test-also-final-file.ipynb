```python
# Function for visualizing the numerical and categorical values

def num_dist(data, var):
    fig, ax = plt.subplots(1, 2, figsize=(12, 4))

    sns.histplot(data=data, x=var, hue='target', palette=['blue', 'orange'], kde=True, ax=ax[0])
    sns.boxplot(data=data, x=var, hue='target', ax=ax[1])
    ax[0].set_title(f"{var} Distribution Histogram")
    ax[1].set_title(f"{var} Distribution Boxplot")

    plt.show()
    
def cat_dist(data, var):
    fig, ax = plt.subplots(1, 2, figsize=(12, 4))

    df[var].value_counts().plot(kind="pie", explode=[0.05 for x in data[var].dropna().unique()], autopct='%1.1f%%', ax=ax[0], shadow=True)
    ax[0].set_title(f"{var} Pie Chart")
    ax[0].set_ylabel('')

    count = sns.countplot(x=var, data=df, hue='target',palette=['blue', 'orange'], ax=ax[1])
    for bar in count.patches:
        count.annotate(format(bar.get_height()),
            (bar.get_x() + bar.get_width() / 2,
            bar.get_height()), ha='center', va='center',
            size=11, xytext=(0, 8),
            textcoords='offset points')
    ax[1].set_title(f"{var} Bar Chart")
    plt.show()
```

Preprocessing

```python
#taking only the first 3 letters in skyconditions_o column 

sky_cond_o=[]
for i in df['SkyConditions_o']:
    a=str(i)
    a=a[:3]
    sky_cond_o.append(a)
```

```python
df['SkyConditions_o']=sky_cond_o
```

```python
#taking only the first 3 letters in skyconditions_d column 

sky_cond_d=[]
for i in df['SkyConditions_d']:
    a=str(i)
    a=a[:3]
    sky_cond_d.append(a)
```

```python
df['SkyConditions_d']=sky_cond_d
```

```python
# taking only the numerical value from visibility_o column

vis_o=[]
for i in df['Visibility_o']:
    a=str(i)
    b=a.split('S')
    c=b[0]
    vis_o.append(c)
```

```python
df['Visibility_o']=vis_o
```

```python
# taking only the numerical value from visibility_d column

vis_d=[]
for i in df['Visibility_d']:
    a=str(i)
    b=a.split('S')
    c=b[0]
    vis_d.append(c)
```

```python
df['Visibility_d']=vis_d
```

```python
# substituting VRB in WindDirection_o column with the median of 60-360

wind_o=[]
for i in df['WindDirection_o']:
    if i=='VRB':
        a='210'
    else:
        a=i
    wind_o.append(a)
```

```python
df['WindDirection_o'] = wind_o 
```

```python
# substituting VRB in WindDirection_d column with the median of 60-360

wind_d=[]
for i in df['WindDirection_d']:
    if i=='VRB':
        a='210'
    else:
        a=i
    wind_d.append(a)
```

```python
df['WindDirection_d']=wind_d
```

```python
df['WindDirection_d']=df['WindDirection_d'].astype(float)
```

```python
df['WindDirection_o']=df['WindDirection_o'].astype(float)
```

- Month 1 had the least number of flights flying, but comparatively had a lot of delayed flights.
- Month 7 had the most flights arriving and departing
- Months 1 and 11 had the highest ratio of delayed flights

- A lot of delays happened in between 0 to 900 seconds. Which explains why the dataset is imbalanced as we have 
  taken delays under 900seconds (15mins) as no delay (2)
    
- Ground Height, Station Height, Barometer Height are very similar columns as they all measure the height of the airport
  from the ground level. Hence one of these columns has to be kept and the others should be removed. This holds true
  for both the origin and destination airports

- Visibility at the origin airport >=7, very less delays
- Visibility at the destination airport >=9, very less delays
Hence we can conclude that Visibility can be taken as one of the major factors for contributing to flight delays

Using SMOTE (Synthetic Minority Oversampling Technique) technique, to make the dataset balanced

```python
from imblearn.combine import SMOTETomek
```

```python
from collections import Counter
```

```python
os=SMOTETomek(1)
x_os,y_os=os.fit_resample(x,y)
print("The number of classes before fit {}".format(Counter(y)))
print("The number of classes after fit {}".format(Counter(y_os)))
```

train-test split

```python
xtrain, xtest, ytrain, ytest = train_test_split(x_os,y_os,test_size=0.3, stratify=y_os, random_state=123)
```
