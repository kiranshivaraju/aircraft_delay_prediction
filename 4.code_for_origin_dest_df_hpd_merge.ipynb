```python
import csv

with open('hpd1.csv', 'r') as file:
    reader1 = csv.reader(file)
    data1 = list(reader1)
    
with open('hpd2.csv', 'r') as file:
    reader2 = csv.reader(file)
    data2 = list(reader2)

with open('hpd3.csv', 'r') as file:
    reader3 = csv.reader(file)
    data3 = list(reader3)
    
with open('hpd4.csv', 'r') as file:
    reader4 = csv.reader(file)
    data4 = list(reader4)
    
with open('hpd5.csv', 'r') as file:
    reader5 = csv.reader(file)
    data5 = list(reader5)
    
with open('hpd6.csv', 'r') as file:
    reader6 = csv.reader(file)
    data6 = list(reader6)
```

```python
origin_data_hpd=[]
count=0
for i in df['origin_neww']:
    i=str(i)
    a=i[4:6]
      
    if a =='01':
        data = data1
    elif a =='03':
        data = data2
    elif a == '05':
        data = data3
    elif a =='07':
        data = data4
    elif a == '09':
        data = data5
    else :
        data = data6
        
    target_value = int(i)
    column_index = 5
    closest_value = None
        
    for row in data[1:]:           
        value = int(row[column_index])
        if closest_value is None or abs(value - target_value) < abs(closest_value - target_value):
            closest_value = value
            row_info = row

    count=count+1
    origin_data_hpd.append(row_info)
    
    print(count)
```
