## Preprocessing

```python
### Converting the ActualArrivalTimeStamp to the required DateTime dtype
df['ActualArrivalTimeStamp'] = pd.to_datetime(df['ActualArrivalTimeStamp'], format='%d/%m/%y %H:%M')
```

```python
df.dtypes
```

```python
#Changing the datatypes of some columns to strings dtype for further preprocssing

doa = df.astype({'DayofMonth':'str', 'Year':'str', 'Month':'str', 'ScheduledArrTime':'str', 'ActualArrivalTimeStamp':'str'})
```

```python
doa
```

### Creating a new column which has the date and time at which the flight was supposed to land

```python
date_list=[]
for i in doa['ActualArrivalTimeStamp']:
    i=str(i)
    split_i=i.split(' ')
    date=split_i[0]
    date_list.append(date)
```

```python
date_list
```

```python
qw=[]
for t in doa['ScheduledArrTime']:
    if len(t)==3:
        a='0'+t
    else:
        a=t
    
    qw.append(a)
```

```python
time_list=[]
for i in qw:
    i_str=str(i)
    time=i_str[0:2]+':'+i_str[2:]
    time_list.append(time)
```

```python
time_list
```

```python
result=[]
for a,b in zip(date_list,time_list):
    c=a+' '+b
    print(c)
    result.append(c)
```

```python
len(result)
```

```python
doa['result']=result
```

```python
doa
```

```python
doa['result'] = pd.to_datetime(doa['result'])
```

### Finding the Dely

```python
#Finding the delay in the arrival time of the flights by subtracting the result and actual arrival time stamp

import datetime
format = '%Y-%m-%d %H:%M:%S'

delay = []

for a, b in zip(doa['result'], doa['ActualArrivalTimeStamp']):
    startDateTime = datetime.datetime.strptime(str(a), format)
    endDateTime = datetime.datetime.strptime(str(b), format)
    diff = endDateTime - startDateTime
    delay.append(diff)
```

```python
delay
```

```python
doa['delay']=delay
```

```python
doa.head()
```

```python
doa['delay'].value_counts()
```

```python
### Creating the Flight Delay Status column
### Flight delayed if delay is greater than 15min, else not delayed

label=[]
time_delay_list=[]
for i in doa['delay']:
    i_str = str(i)
    split_t = i_str.split(' ')
    days=int(split_t[0])
    time=split_t[-1]
    time_split=time.split(':')
    hours=int(time_split[0])*3600
    mins=int(time_split[1])*60
    secs=int(time_split[2])
    
    if days>=0:
        time_delay = days+hours+mins+secs
    else:
        time_delay = 0
    
    time_delay_list.append(time_delay)
    
    if time_delay>900:
        label.append(1)
    else:
        label.append(2)
```

```python
label
```

```python
doa['delay_in_seconds']=time_delay_list
doa['target']=label
```

```python
doa
```

```python
### Saving the dataframe to csv

doa.to_csv('train_with_target.csv')
```
