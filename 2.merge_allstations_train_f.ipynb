## Merging train dataset with all stations dataset to get the weather station IDs of the origin and departure airports

Importing the datasets and required librares

```python
import pandas as pd
```

```python
train=pd.read_csv('C:/Users/kiran/OneDrive/Desktop/final/train_with_target.csv')
```

```python
train
```

```python
train = train.drop('Unnamed: 0', axis=1)
```

```python
train
```

```python
allstationsdf=pd.read_csv('C:/Users/kiran/OneDrive/Desktop/hacakthon_final/ProblemStatement_Data&Instructions-1616503373207/DataForPHD_Forall/20210322_weatherdatatoconsider/AllStationsData_PHD.txt', delimiter='|')
```

```python
allstationsdf
```

### Merging the datasets which have the same columns

```python
### Merging together the train dataset with the allstationsdf and matching it with origin AirportID

origin_df=pd.merge(train, allstationsdf, left_on='Origin', right_on='AirportID')
```

```python
origin_df.head()
```

```python
origin_df
```

```python
### Merging together the train dataset with the allstationsdf and matching it with departure AirportID

dest_df=pd.merge(train, allstationsdf, left_on='Destination', right_on='AirportID')
```

```python
dest_df.head()
```

```python
dest_df
```

```python
### Combining the 2 datasets to make it one

merged=pd.merge(origin_df,dest_df, on='FlightNumber',how='inner')
```

```python
merged
```

```python
merged.columns
```

```python
train.columns
```

### Removing the repeated columns

```python
merged=merged.drop(['Year_y', 'Month_y', 'DayofMonth_y', 'DayOfWeek_y',
       'ScheduledDepTime_y', 'ScheduledArrTime_y', 'ScheduledTravelTime_y',
       'Origin_y', 'Destination_y', 'Distance_y', 'ActualArrivalTimeStamp_y', 'result_y', 'delay_y',
                   'delay_in_seconds_y', 'target_y'],axis=1)
```

```python
merged.columns
```

```python
### Renaming the columns

merged.columns=['FlightNumber', 'Year', 'Month', 'DayOfMonth', 'DayOfWeek',
       'ScheduledDepTime', 'ScheduledArrTime', 'ScheduledTravelTime',
       'Origin', 'Destination', 'Distance', 'ActualArrivalTimeStamp',
        'result', 'delay', 'delay_in_seconds', 'target',
       'WeatherStationID_origin', 'AirportID_origin', 'GroundHeight_origin',
       'StationHeight_origin', 'BarometerHeight_origin', 'Latitude_origin', 'Longitude_origin',
       'TimeZone_origin', 'WeatherStationID_dest', 'AirportID_dest', 'GroundHeight_dest',
       'StationHeight_dest', 'BarometerHeight_dest', 'Latitude_dest', 'Longitude_dest',
       'TimeZone_dest']
```

```python
### Saving the dataframe into a csv file

merged.to_csv('merged_with_target_allstations.csv')
```
